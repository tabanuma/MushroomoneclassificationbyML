```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
print(y)
print('\n')
print(X[:3])
```

```python
# splitting into training and testing sets
from sklearn.model_selection import train_test_split

# default split ratio is 75:25
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=11)
X_train[:5]
```

```python
# viewing y test
y_test
```

```python
# Encoding the target value
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()

y_train = labelencoder.fit_transform(y_train)
print(y_train)

y_test = labelencoder.fit_transform(y_test)
print(y_test)
print(len(y_train), len(y_test))
```

```python
# Encoding the set of X predictors using one hot encoding
from sklearn import preprocessing
encoder = preprocessing.OneHotEncoder()
encoder.fit(X_train)

X_train = encoder.transform(X_train).toarray()
print(X_train)

X_test = encoder.transform(X_test).toarray()
print(X_test)

print(len(X_train), len(X_test))
```

```python
# creating a dataframe to capture metrics for each model

model_metrics = pd.DataFrame(columns=['model', 'accuracy_train', 'accuracy_test', 'avg_cross_val_score', 'recall_train', 'recall_test',
                                     'precision_train', 'precision_test', 'f1_train', 'f1_test', 'auc_train', 'auc_test', 'saved_model'])
model_metrics
```

```python
# list of all categorical models from sklearn:
from sklearn.utils import all_estimators

estimators = all_estimators(type_filter='classifier')
for name, class_ in estimators:
    module_name = str(class_).split("'")[1].split(".")[1]
    class_name = class_.__name__
    print(f'from sklearn.{module_name} import {class_name}')
```

```python
# Experimenting with various models out of the box (without hyper-parameter tuning)
# to find the most appropriate one to tune further

# importing metrics
from sklearn import metrics
from sklearn.metrics import recall_score, precision_score, accuracy_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.calibration import CalibratedClassifierCV

# importing models
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import Perceptron
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from xgboost.sklearn import XGBClassifier


models = [DummyClassifier(), LogisticRegression(), SVC(probability=True), GaussianNB(), MultinomialNB(), CategoricalNB(), SGDClassifier(),
          Perceptron(), KNeighborsClassifier(), DecisionTreeClassifier(), RandomForestClassifier(),
          GradientBoostingClassifier(), LGBMClassifier(verbose= -100), XGBClassifier()
         ]


# using i for the df index
i = 0
# creating a dictionary to store data for ROC plots
roc_data = {}

for model in models:
    # storing the cross val score first
    cross_val_accuracy = cross_val_score(estimator=model, X=X_train, y=y_train, cv = 10).mean()

    # fittting the model on the training set and generating predictions on the training and testing sets
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # predicting probabilities for an roc curve
    # Note: SGDClassifier() can't predict probabilities with loss of hinge, but we can pass it to CalibratedClassifier to do this
    if str(type(SGDClassifier())) == "<class 'sklearn.linear_model._stochastic_gradient.SGDClassifier'>":
        calibrator = CalibratedClassifierCV(model, cv='prefit')
        calibrator.fit(X_train, y_train)
        y_pred_prob_train = calibrator.predict_proba(X_train)[:,1]
        y_pred_prob_test = calibrator.predict_proba(X_test)[:,1]
    else:
        y_pred_prob_train = model.predict_proba(X_train)[:, 1]
        y_pred_prob_test = model.predict_proba(X_test)[:, 1]
    # calcing metrics
    saved_model = model
    model = str(model).split('(', 1)[0]
    acc_train = accuracy_score(y_train, y_pred_train)
    acc_test = accuracy_score(y_test, y_pred_test)
    recall_train = recall_score(y_train, y_pred_train, zero_division=1)
    recall_test = recall_score(y_test, y_pred_test, zero_division=1)
    precision_train = precision_score(y_train, y_pred_train, zero_division=1)
    precision_test = precision_score(y_test, y_pred_test, zero_division=1)
    f1_train = f1_score(y_train, y_pred_train, zero_division=1)
    f1_test = f1_score(y_test, y_pred_test, zero_division=1)

    # using probabilities for the roc_auc_score
    auc_train = roc_auc_score(y_train, y_pred_prob_train)
    auc_test = roc_auc_score(y_test, y_pred_prob_test)
    # saving data to plot roc curves later
    roc_data[model] = [auc_test, y_pred_prob_test]

    # adding metrics to results df
    model_results = [model, acc_train, acc_test, cross_val_accuracy, recall_train, recall_test, precision_train, precision_test,
                    f1_train, f1_test, auc_train, auc_test, saved_model]
    model_metrics.loc[i] = model_results
    i += 1



model_metrics.sort_values(by=['accuracy_test', 'avg_cross_val_score'], ascending=False)
```

```python
# confusion matrix for random forest
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

best_model = model_metrics.loc[model_metrics['model'] == 'RandomForestClassifier']['saved_model'].values[0]
y_pred = best_model.predict(X_test)


conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot = True, linewidths=0.2, linecolor="black", fmt = ".0f",
            cmap="Blues", xticklabels=["Edible", "Poisonous"], yticklabels=["Edible", "Poisonous"])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title('Confusion Matrix: Random Forest Classifier ')
plt.show()
```

```python
# plotting the ROC Curve for all classifiers to compare models

from matplotlib.pyplot import figure
#figure(figsize=(8, 6), dpi=80)
figure(figsize=(10, 8), dpi=80)

for model in roc_data:
    auc_score = round(roc_data[model][0],4)
    predicted_prob = roc_data[model][1]
    fpr, tpr, _ = metrics.roc_curve(y_test, predicted_prob)
    plt.plot(fpr, tpr, label=model + ", AUC=" + str(auc_score))

plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve for all Classifiers')
plt.legend(bbox_to_anchor=(1.42, 1.01), loc='upper right', ncol=1)
plt.show()
```
